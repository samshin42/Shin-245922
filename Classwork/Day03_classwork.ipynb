### 2.  Loops

The code cell below contains a `for` loop.  We first initialize the value of `S` to be 0.  Then we compute a running total inside the loop.  

The built-in function `range` is effectively a way of quickly creating a sequence of numbers.  The inputs for `range(a, b, c)` are as follows:  `a` is the starting value, `b` is the stopping value plus one, and `c` is the step size.  So `range(1,101,1)` produces the sequence of numbers 1, 2, 3, ..., 99, 100.  You can see the final value of `n` produced by the first print statement outside the for loop.

Note the way that `S += n` is indented, while `print(n)` and `print(S)` are not indented.  These indentation levels are very important in Python code.  See what happens if you move the print statements inside the for loop by changing their indentation level to match that of `S += n`.

````python
'''
Created by: CAPT Foraker
Date Created: 08/19/2019
Date Edited: 12/20/2019

This code contains a for loop that will sum the first 100 natural numbers.
'''

# Create a variable, S, to store the summed value
S = 0

# It's a for loop
for n in range(1, 101, 1):
    # Add the number n to the running total S in the "Pythonic" way
    S += n
    # The line of code above is equivalent to S = S + n, but that is not the "Pythonic" way to write it
    
print(n)
print(S)
````

Another way to loop in Python is illustrated below.  This time we loop over a collection of items in a list.  The key idea of the **for** loop is that it allows you to perform the same task or set of tasks on every item in a list.

````python
'''
Created by: CAPT Foraker
Date Created: 08/19/2019
Date Edited: 12/20/2019

This code contains another example of a for loop over a collection of items in a list.
'''

# A list of the names of the members of The Band
TheBand = ['Robbie Robertson', 'Levon Helm', 'Richard Manuel', 'Rick Danko', 'Garth Hudson']

# Print each member's name
for name in TheBand:
    print(name)

# How do we write code outside the for loop?
print("And those are the members of The Band!")
````

### 3.  Slicing

Slicing is useful if you have a data structure like a list and you would like to obtain specific sub-elements from it, without using a more complicated structure like a loop.  The code block below illustrates slicing by starting with a list called `my_list` that contains the numbers 1 through 8.

The syntax for slicing a list is as follows:  `my_list[a:b:c]`, where `a` is the position in the list where you would like to start slicing, `b` is one more than the position in the list where you would like to stop slicing, and `c` is the increment you would like to use when doing the slicing (the default increment is one in Python).

````python
my_list = [1,2,3,4,5,6,7,8]
start = 0
stop_plus_one = 8
increment = 1
# Note that the three lines of code below ALL produce the same output
print(my_list[start:stop_plus_one:increment])
print(my_list[::])
print(my_list[:])
````

````python
# Use slicing to only get the first 4 elements of my_list
print(my_list[0:4])
# Use slicing to TEMPORARILY reverse my_list
print(my_list[::-1])
print("Note that my_list[::-1] did not permanently modify my_list")
print(my_list)
# Use .reverse() to PERMANENTLY reverse my_list
my_list.reverse()
print("Note that my_list.reverse() PERMANENTLY modifies my_list")
print(my_list)
````

### Slicing Summary

- Slicing is like indexing on steroids. 🤪

- A __slice__ is a specific group of items in a list.

- `my_list[start:stop]` gives you a new list containing the elements of `my_list` whose indices start at `start` and end at `stop - 1`.

- `my_list[start:stop:step]` gives you a new list containing the elements of `my_list` whose indices start at `start`, and increase by `step` until `stop - 1` is reached or passed.

- If you omit `start`, Python automatically starts your slice at the beginning of `my_list`

- If you omit `stop`, Python automatically ends your slice at the end of `my_list`

- If you omit `step`, Python automatically assumes a step value of 1.

````python
import matplotlib.pyplot as plt
````

* `as plt` in the cell above lets us refer to `matplotlib.pyplot` as `plt`.
    - This helps us save some keystrokes and keeps our code a bit cleaner.

- The code cell below produces a plot of the parabola $y = x^2$ using Matplotlib.  
    - Sometimes, you have to run the first cell with Matplotlib code twice in your Jupyter notebook in order for the image to appear.

````python
# Create a figure object
fig = plt.figure()

# Draw an axis object on the figure
# Subplot: (rows, cols, item)
ax = fig.add_subplot(1, 1, 1)

# Create list of values on the x-axis between 0 and 5, with an increment of 0.1
x_values = []
for i in range(51):
    x_values.append(0.1 * i)

# Create list of corresponding y values
y_values = []
for x in x_values:
    y_values.append(x**2)

# Plot the data
plt.plot(x_values, y_values)

# Label the axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Give the plot a title
plt.title('A Parabola Using Matplotlib')

# Show the figure
plt.show()
````

- Some notes about the above code:
    - `x_values` and `y_values` are lists of the same size.
    - `plt.plot(x_values, y_values)` plots the points
            
        (`x_values[0]`, `y_values[0]`), (`x_values[1]`, `y_values[1]`),  (`x_values[2]`, `y_values[2]`), ...
        
        and connects all these points (by default). You can change this behavior (connected or not connected, colors, shapes, etc.): see [the documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html).
    - `ax.set_xlabel()` and `ax.set_ylabel()` label the horizontal and vertical axes, respectively. 
    - `plt.title()` sets the plot title.

 - Below is a second example that includes two plots in a single figure.

````python
# Create a figure object
fig = plt.figure()

# Add overall title for the figure
# suptitle is supreme title
fig.suptitle('Example with two plots')

# Create some space between the plots
fig.subplots_adjust(hspace=0.4, wspace=0.4)   

# Draw one axis object on the figure
ax = fig.add_subplot(1, 2, 1)

# Create list of values on the x-axis between 0 and 5, with an increment of 0.1
x_values = []
for i in range(51):
    x_values.append(0.1 * i)

# Create list of corresponding y values
y_values = []
for x in x_values:
    y_values.append(x**2)

# Plot the data
plt.plot(x_values, y_values)

# Label the axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Give the plot a title
plt.title('A Parabola Using Matplotlib')

# Draw a second axis object on the figure
ax = fig.add_subplot(1, 2, 2)

# Create list of values on the x-axis between 0 and 5, with an increment of 0.1
x_values = []
for i in range(51):
    x_values.append(0.1 * i)

# Create list of corresponding y values
y_values = []
for x in x_values:
    y_values.append(x**3)

# Plot the data
plt.plot(x_values, y_values, color='r')

# Label the axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Give the plot a title
plt.title('A Cubic Function Using Matplotlib')

# Show the figure
plt.show()
````
