## 3. Data Types

 - Numbers
     - Integers and Floating-Point Numbers
 - Strings
     - Sequences of character data
     - May be delimited using either single or double quotes
     - All the characters between the opening delimiter and matching closing delimiter are part of the string
     - May contain numbers, letters, symbols, or nothing
     
 - Lists
     - A collection of items that are organized in a particular order
     - Written as a sequence of comma-separated items between square brackets, [...]
     - Can contain numbers, strings, other lists, or nothing
     
 - Tuples
     - Look just like a list except they uses parentheses, ( ), instead of square brackets [ ]
     - Are immutable (items in the tuple **cannot** change)
     - Technically defined by the presence of a comma; the ( ) make them look neater and more readable
     - Can contain numbers, strings, lists, other tuples, or nothing

````python
# Example of integer
print(type(7))
# Example of floating-point number
print(type(2.3))
````

````python
# Examples of strings
print(type('My name is Inigo Montoya, you killed my father, prepare to die!'))
print(type("111000101"))
print(type("1337h4x0r$"))
print(type(""))
````

### Yet another important aside: rules for naming variables

- Variable names can only contain letters, numbers, and underscores.

- Variable names <span class="rred">cannot</span> start with a number.

- Spaces are not allowed! Use underscores instead.

- <font color="red">Avoid</font> using Python keywords and function names &mdash; these are words that Python has reserved for a particular programmatic purpose.
    - For example, don't define a variable called `print` &mdash; how will Python know if you want to use the `print()` function or your variable called `print`?
    - [List of Python keywords](https://www.w3schools.com/python/python_ref_keywords.asp)

- Variable names should be short but descriptive.
    - `name` is better than `n`
    - `student_name` is better than `s_n`
    - `name_length` is better than `length_of_persons_name`

## 7. f-Strings

 - The preferred way to format strings since Python 3.6
 - You can read all about them [here](https://www.python.org/dev/peps/pep-0498/)
 - Below are two examples of how to insert the value of a variable into a string.


````python
my_test = 75
# An OLD way to print the string and numerical value
print("My test score is " + str(my_test))
# The BETTER way to print the string and numerical value
print(f"My test score is {my_test:.2f}")
````

* To insert a variable's value into a string using an f-string:
    - place the letter `f` immediatley before the opening quotation mark
    - put braces around the names of any variables you want to use inside the string
    
* These strings are called __f-strings__
    - "f" is for "format"

\begin{equation*}
\frac{h}{2}\left(1f(x_0)+2f(x_1)+2f(x_2)+\cdots+2f(x_{n-1}) + 1f(x_n)\right),
\end{equation*}
where $h = (b-a)/n$ and $x_i = a+ih$ $(i=0,\ldots,n)$.
